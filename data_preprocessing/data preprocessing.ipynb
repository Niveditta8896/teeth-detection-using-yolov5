# 2 now we r going to shuffle the images before splitting so that we get a fair random split

```python
import random
```

```python
total_images=[]
for folder in image_folder:
    
    all_images=[image for image in os.listdir(folder) if image.endswith((".jpg", ".jpeg", ".png", ".JPG", ".JPEG", ".PNG", ".webp", ".bmp"))]
    random.shuffle(all_images)
    for img in all_images:
        total_images.append(os.path.join(folder, img))#so total_images is now a list of file paths from all the 4 class folders.
    
# this is same as 

# all_images=[]
#for folder in image_folder:
    #for image in os.listdir(folder):
        #if image.endswith((".jpg",".png")):
            
            #all_images.append(image)
        
#random.shuffle(all_images)
```

# 3 now we r going to split the images to train and test set. how? manually and not using sklearn cos its only for classification.

```python
# we r going to split the images like 80:20 as always

split=int(0.80*len(total_images)) # here split is the 80 percent of len(all_images)

# if len(all_images)=200, then split =int(200*.80)=160
```

```python
train=total_images[:split]# for the same eg all_images[all the images upto 160] and are file paths of images
test=total_images[split:]# for the same eg all_images[all the images from 160 to 200]and are file paths of images
```

# 4. after splitting we r going to split the labels to the train and the test images

```python

# we know that the labels of each image will hv the same name as the mage jus a .txt extension will be there.
labels_train=[]
for image in train:
    base = os.path.splitext(os.path.basename(image))[0]
    label_name = base + ".txt"
    
# so our goal is to copy and paste the test image paths to the destination folder same with labels but we hv to create the pathe 
# by joing the label name.txt with label_folder path. y? shutil.copy an dshutil.move only works with file paths.
    shutil.copy(image, os.path.join(output_base_train, os.path.basename(image)))
    shutil.copy(os.path.join(label_folder,label_name),os.path.join(output_base_train_labels,label_name))
```

```python
# this is for test images and respective label names

labels_test=[]
for image in test:
    base = os.path.splitext(os.path.basename(image))[0]
    label_name = base + ".txt"
    
    # Always copy the image
    shutil.copy(image, os.path.join(output_base_test, os.path.basename(image)))
    shutil.copy(os.path.join(label_folder,label_name),os.path.join(output_base_test_labels,label_name))
```
